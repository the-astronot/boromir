```python
def single_pose(state,sun_los):
	text = "#ImgName,CAM_PosX(m),CAM_PosY(m),CAM_PosZ(m),CAM_QuatS,CAM_QuatV1,CAM_QuatV2,CAM_QuatV3,Time,SUN_LosX,SUN_LosY,SUN_LosZ,EARTH_PosX(m),EARTH_PosY(m),EARTH_PosZ(m),EARTH_QuatS,EARTH_QuatV1,EARTH_QuatV2,EARTH_QuatV3\n"
	data_len = 19
	for i,los in enumerate(sun_los):
		data = ["" for _ in range(data_len)]
		data[0] = "image{}".format(i)
		data[1:4] = state.position.tolist()
		data[4] = state.attitude.s
		data[5:8] = state.attitude.v.tolist()
		data[9:12] = los.tolist()
		for i in range(data_len):
			data[i] = str(data[i])
		text += "{}\n".format(",".join(data))
	print(text)
```
